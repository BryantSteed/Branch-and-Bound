```python
import numpy as np
import math
from tsp_solve import branch_and_bound
from tsp_core import Timer
```

```python
cost_matrix = [
        [math.inf, 20, 30, 10, 11],
        [15, math.inf, 16, 4, 2],
        [3, 5, math.inf, 2, 4],
        [19, 6, 18, math.inf, 3],
        [16, 4, 7, 16, math.inf]
    ]

result = branch_and_bound(cost_matrix, Timer(20))
print("Best tour found:", result[-1].tour)
print("Cost of the best tour:", result[-1].score)
```
