```python
from tsp_solve import branch_and_bound
from performance import run_analysis
from tsp_core import generate_network, Timer
from copy import deepcopy
```

```python
def perf_test_func(size: int) -> bool:
    locations, edges = generate_network(
        size,
        euclidean=True,
        reduction=0.2,
        normal=False,
        seed=312,
    )

    timer = Timer(10000)
    stats = branch_and_bound(deepcopy(edges), timer)

    if stats and not timer.time_out():
        final_score = stats[-1].score
        print(f"Size: {size}, Final Score: {final_score}")
        return True
    return False
```

```python
sizes = [5, 10, 15]
run_analysis(sizes, perf_test_func)
```
